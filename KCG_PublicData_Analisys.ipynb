## 해양경찰청 해상단속 현황

최근에는 공공기관에서 연간, 혹은 일정 기간의 자료들을 국민에게 '공공데이터'로서 공개하고 있다. '청렴'을 모토로 삼고 있는 해양경찰 또한 국민이 해경의 임무 수행과 활약에 대해 알 수 있도록 공공데이터 개방을 실시하고 있다. 오늘날 방대한 양의 데이터 속에서 이러한 지표는 해경과 같은 기관이나 단체, 더 나아가 사회를 이해하기 위해 필수적인 조건이라고 생각한다.<br>
해양경찰 국민기자단으로 활동하며 해경을 가까이서 접했고, 1년의 시간 동안 해경의 현장과 여러 요소들을 경험했다. 현장에서는 해경의 임무 수행과 작업을 생동감 있는 모습으로 체험했고, 간접적으로도 해양경찰이 국민의 삶 속에서 어떻게 역할을 하는 지 알게 되었다.<br>
대표적인 해양경찰의 임무는 해상에서 발생하는 사건들의 단속이다. 단속을 하며 범죄 현장을 검거하기도 하고, 불법조업 어선을 나포하여 인계하기도 한다.<br>
이 분석에서는 이러한 현황을 나타낸 기본 데이터를 불러오고, 그래프로 시각화함으로써 데이터 간의 관계성을 더 용이하게 파악해보려 한다.

```python
import pandas as pd
import numpy as np
from pandas import DataFrame
```

아래의 데이터는 **해경의 해상 방범 활동에 따라 파악한 해상범죄 발생 및  단속 현황**이다. 범죄 발생에 따라 해경에서 검거한 현황을 최근 10년 간의 지표로 자세하게 보여주고 있다.

```python
df_maritime_regulation = pd.read_csv('C:\\Users\\LG\Desktop\\KCG_Data_Analisys\\해양경찰청_해상방범활동 해상범죄 발생 및 단속 현황.csv', encoding='cp949')
df_maritime_regulation.set_index('기준 연도', inplace = True)
df_maritime_regulation
```

다음 데이터는 **불법조업 외국어선 나포 현황** 이다. 위의 해상범죄 단속 현황을 통해 해경의 주된 임무 중 하나인 불법조업 단속은 어떤 동향을 나타내고 있는 지 궁금증을 가졌다. 아래 데이터를 살펴보면, 불법조업 단속 현황이 월별로 표시되어 있다. '16년도부터 가장 최근인 '21년도까지 살펴봤을 때, 불법조업 현황이 대폭 감소하였다.'19년도 이전의 불법조업은 한 해 동안 꾸준히 존재했고, 특히 봄과 가을/겨울에 많이 나타났다. 하지만 그의 정도가 대폭 감소한 '19 이후에도 불법조업은 여전히 봄과 가을/겨울 철에 집중되어 있는 것을 볼 수 있다.

```python
df_illegal_fishing = pd.read_csv('C:\\Users\\LG\Desktop\\KCG_Data_Analisys\\해양경찰청_불법조업 외국어선 월별 나포 현황.csv', encoding='cp949')
df_illegal_fishing = df_illegal_fishing.fillna('0')
df_illegal_fishing = df_illegal_fishing.sort_values(by=['년도_월'] ,ascending=True)
df_illegal_fishing.set_index("년도_월", inplace = True)
df_illegal_fishing
```

마지막 지표는 **중국어선 조업 현황**이다. 대한민국 해양경찰이 대표적으로 집중 단속하는 불법조업 외국어선은 대부분 중국에서 넘어오는 경우가 많다. 따라서 중국어선이 해경의 관할 구역에서 조업하는 현황을 가져와 보았다. 그 현황은 '허가수역'과 *'NLL해역'* 으로 나눌 수 있는데, NLL해역에서는 그 수가 많이 감소했지만, 허가수역에서는 100에서 200건을 유지하는 모습이다.<br>
***NLL: 북방한계선***

```python
df_chinese_vessel = pd.read_csv('C:\\Users\\LG\Desktop\\KCG_Data_Analisys\\해양경찰청_중국어선 조업현황.csv', encoding='UTF-8')
df_chinese_vessel.set_index("년도", inplace = True)
df_chinese_vessel = df_chinese_vessel.T
df_chinese_vessel
```

위의 3가지 그래프를 시각화 해보았다. 위와 같은 표로 보아도 각 시기에 알맞은 현황을 파악하는 데에는 아무런 지장이 없지만, 데이터 파악을 조금 더 용이하게 하고자 Matplot 라이브러리를 이용하여 데이터를 그래프로 표현하였다.

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import platform

from matplotlib import font_manager,rc
plt.rcParams['axes.unicode_minus']=False

if platform.system()=='Darwin':
    rc('font',family='AppleGothic')
elif platform.system()=='Windows':
    path='c:/Windows/Fonts/malgun.ttf'
    font_name=font_manager.FontProperties(fname=path).get_name()
    rc('font',family=font_name)
else:
    print('Unknown system')
```

```python
for col in df_maritime_regulation.columns.to_list():
    df_maritime_regulation[col] = df_maritime_regulation[col].astype('float')
plt.figure(figsize=(15,10))
plt.grid()
plt.xlabel("연도")
plt.ylabel("건수")
plt.title("해상방범활동 해상범죄 발생 및 단속 현황")

for col in df_maritime_regulation.columns.to_list():
    plt.plot(df_maritime_regulation.index, df_maritime_regulation[col], marker='o')
    
plt.legend(df_maritime_regulation.columns)

plt.show()
```

위 데이터 프레임에서 본 것처럼 발생 건수와 검거 건수는 거의 같은 움직임을 보이고 있다. 데이터 프레임에 나타난 것과 같이 해경의 검거율은 매번 90% 이상을 기록하고 있기 때문이다.<br>
하지만, 위 그래프에서는 10000 단위의 지표가 나오기 때문에 정작 100% 대의 범위가 기록되어야 하는 '검거율'과, 100에서 200 사이의 범위를 보이는 '구속수'는 제대로 표현되기가 어려웠다.<br>
따라서 두 지표를 따로 분리한 뒤, y축 간격을 좁혀서 아래에서 살펴보았다.

```python
df_maritime_regulation["검거율"] = df_maritime_regulation["검거율"].astype('float')
df_maritime_regulation["구속수"] = df_maritime_regulation["구속수"].astype('float')

plt.figure(figsize=(10,5))
plt.grid()
plt.xlabel("연도")
plt.ylabel("건수")
plt.title('해상범죄 검거율&구속수')

plt.plot(df_maritime_regulation.index, df_maritime_regulation["검거율"], '#d62728', marker='o')
plt.plot(df_maritime_regulation.index, df_maritime_regulation["구속수"], '#9467bd', marker='o')
    
plt.legend(['검거율', '구속수'])

plt.show()
```

이제 검거율과 구속수는 더욱 지표 파악에 용이하게 되었다. 위에서 파악했던 것처럼 검거율은 항상 100%에 가까웠고, 구속수는 '12년도와 '13년도에 상승했다가 이후로 점차 감소하고 있다.

```python
for col in df_illegal_fishing.columns.to_list():
    df_illegal_fishing[col] = df_illegal_fishing[col].astype('float')
plt.figure(figsize=(13,8))
plt.grid()
plt.xlabel("연도")
plt.ylabel("월별 건수")
plt.title("불법조업 외국어선 월별 나포 현황")

for col in df_illegal_fishing.columns.to_list():
    plt.plot(df_illegal_fishing.index, df_illegal_fishing[col], marker='o')
    
plt.legend(df_illegal_fishing.columns)

plt.show()
```

위 그래프는 불법조업 외국어선 월별 나포 현황이다. 하지만 12개의 여러 데이터가 겹쳐 있다보니 파악하기에 다소 복잡한 모양을 보인다. 따라서 subplot() 함수를 통해 그래프를 월별로 분리해 보았다.

```python
col = df_illegal_fishing.columns.to_list()
color = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf', '#1f77b4', '#ff7f0e']
sub_plots = plt.subplots(3, 4, figsize=(15,10))

fig = sub_plots[0]
graph = sub_plots[1]

for i in range(3):
    for j in range(4):
        graph[i][j].plot(df_illegal_fishing.index, df_illegal_fishing[col[(i * 4) + j]], color = color[(i * 4) + j])
        graph[i][j].set_title(col[(i * 4) + j])
        graph[i][j].set_ylim([0, 50])
plt.show()
```

그래프를 월별로 따로 분리하니 더욱 파악하기 용이했다. 그래프는 4,5월 지표와 9월~12월 지표가 상대적으로 높아 보인다. 이는 데이터 프레임에서 살펴보았던 같과 다를 바 없이, 불법조업 외국어선은 봄철과 가을/겨울철에 집중되어 있다는 것을 나타낸다.

```python
for col in df_chinese_vessel.columns.to_list():
    df_chinese_vessel[col] = df_chinese_vessel[col].astype('float')
plt.figure(figsize=(10,5))
plt.grid()
plt.xlabel("연도")
plt.ylabel("연도별 건수")
plt.title("중국어선 조업현황")

for col in df_chinese_vessel.columns.to_list():
    plt.plot(df_chinese_vessel.index, df_chinese_vessel[col], linestyle='dashdot', marker='o')
    
plt.legend(df_chinese_vessel.columns)

plt.show()
```

마지막으로 중국어선 조업 현황이다. 허가수역에서의 조업 현황은 NLL에서보다 항상 높은 수치였다. 하지만 그래프를 통해서 본 전체적인 움직임은 매우 달랐다. 허가수역 내에서의 조업은 '19년도를 기점으로 감소를 보이는 반면, NLL에서의 조업은 '18년도를 기준으로 꾸준하게 증가하고 있었다.<br>
이처럼 일반 데이터 프레임에서는 자세하게 파악하지 못했던 부분들이 데이터를 그래프로 시각화할 때 보였고, 지표의 움직임 분석에도 상대적으로 용이했다.
