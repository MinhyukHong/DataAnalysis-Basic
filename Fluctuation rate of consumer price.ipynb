## - 국제유가와 원/달러 환율을 중심으로

## 주제 선정 이유

최근 인플레이션이 가속화되고 소비자물가가 상승하고 있다. 물가 상승의 원인으로는 코로나19에 따른 경기 하강에 대응하기 위한 시중 통화량 증가, 국제 유가의 폭등, 원/달러 환율 상승이 있다.<br />
오늘날 우리는 방대한 양의 데이터 속에서 살아가고 있다 해도 과언이 아니다. 그렇기에 세상을 표면적으로만 보는 것이 아닌, 쏟아지는 데이터를 통한 정보의 수집과 그 원인의 분석은 삶을 살아가는 데에 있어 하나의 현상을 다양한 시각과 해석으로 접근하게 해준다. 최근 가파른 물가 상승이 계속되면서 다양한 영역이 타격을 받았다. 고도의 정보화 사회인 오늘날 물가 상승의 원인은 당장 검색을 통해 알 수 있을 지라도, 상승 폭의 변화와 상승 요인과의 상관 관계를 체계적으로 분석해본다면 이러한 원인이 우리나라 경제와 실제 삶에 미치는 영향을 구체적으로 알 수 있을 것이라 생각했다. 더 나아가 앞으로의 삶의 현상을 이해하는 데에도 도움이 되고자 이 주제를 선정하였다.
#### 국제 유가 상승
우리나라를 포함한 많은 국가에서 코로나19가 완화되며 세계 경제는 빠르게 회복되었고 석유 사용량의 증가를 불러왔다. 더불어 유럽과 미국의 탄소 배출 규제는 탄소 배출이 적은 천연가스의 가격을 상승하게 하여 석유 사용을 더 증가시켰다.석유 사용량이 증가하며 국제유가는 상승했고 석유라는 제한된 자원 안에서 수요와 공급을 맞추기란 쉽지 않은 문제이다. 설상가상으로 러시아와 우크라이나의 전쟁은 국제유가의 상승을 부추겼다. 러시아는 세계 3위 산유국이다. 우크라이나 침공으로 인한 서방 국가들의 제재가 심화하면 글로벌 공급망에 심각한 불균형을 초래할 것이란 우려가 유가 상승으로 이어졌다. 또한 러시아산 원유 수출가격 상한제를 도입하며 러시아의 원유 생산량을 줄이고 국제유가 상승 압력이 가속화되었다. 이로 인한 국제유가의 상승은 원자재 가격의 급등을 불러 일으켰고, 이 여파가 본격적으로 우리나라 경기에 반영되었다.
#### 원/달러 환율 상승
현재 국내를 비롯한 세계는 물가 상승으로 인해 고통 받고 있다. 코로나19 사태를 거치며 그 시기에 우리나라는 부양 정책으로 정부와 중앙은행이 국민들에게 많은 양의 돈을 지급했다. 코로나 지원금 등의 명목으로 지급하게 된 돈이지만, 이 지원금은 국민의 높은 소비를 불러 일으켰고 결과적으로 상품과 서비스의 공급은 그대로인 반면 폭발적인 수요의 증가로 인플레이션이 발생하게 되었다. 시중 통화량의 증가는 화폐가치를 하락하게 하고, 우리나라의 경우 원화 가치가 하락하면 달러화는 원화 대비 안정적인 화폐가 된다. 즉 원화의 약세는 달러화의 강세가 되고 환율의 상승을 야기한다. 또한 수입과 수출 의존도가 높은 우리나라는 상품과 서비스를 해외에서 달러화로 들여오기 때문에 환율의 상승은 물가의 상승까지 이어졌다.

## 가설 정의

국제 유가와 원/달러 환율이 상승함에 따라 물가가 상승한다는 것은 일반적인 사실이다. 하지만 실제 데이터를 통해 상관 관계를 분석한다면 그 동향을 더 쉽게 파악할 수 있을 것이라 판단했다. 본 프로젝트를 통하여 환율과 유가가 변동함에 따라 물가는 어떠한 경향과 상승/하락 폭을 가지고 움직이는 지 알아볼 계획이다. 대한민국 국민으로서, 국제사회의 일원으로서, 유가와 환율, 물가와 같은 국가와 사회를 움직이는 여러 요소들이 어떤 가치를 지니는지 보다 쉽게 파악하기 위해 이 프로젝트가 필요하다고 생각했다.

### 구현 내용 설명

아래의 데이터프레임에는 최근 6년간의 소비자 물가 상승률과 더불어 여러 분야에서의 물가 데이터를 파악할 수 있는 다양한 지표가 있다.

```python
import pandas as pd
import numpy as np
from pandas import DataFrame

Data_Price = pd.read_csv('C:\\Users\\LG\\Desktop\\Python Term_Project\\소비자물가 상승률.csv', encoding="UTF-8")
Data_Price = Data_Price.fillna('')
Data_Price = Data_Price.drop([8, 9], axis = 0)
Data_Price.set_index("물가", inplace=True)
#pd.to_datetime(Data_Price.columns)
col = Data_Price.columns.tolist()
col[5:]= [x.strip('월') for x in col[5:]]
col
Data_Price.columns = col
Data_Price_2022 = Data_Price[["2022 04", "2022 05", "2022 06", "2022 07", "2022 08", "2022 09"]]
Data_Price_2022.mean(axis=1)
Data_Price["2022"] = Data_Price_2022.mean(axis=1)
Data_Price["2022"] = round(Data_Price["2022"], 1)
Data_Price = Data_Price[["2017", "2018", "2019", "2020", "2021", "2022"]]
Data_Price = Data_Price.T
Data_Price
```

약 5년 간의 물가 지표 중 최근 2022년에 가장 많은 상승률을 보인 물가를 왼쪽부터 순서대로 나열해보면 다음과 같다.

```python
Data_Price.sort_values(by='2022', axis=1, ascending=False)
```

공업제품은 2022년에 8% 대로 가장 높은 상승률을 보이고 있다.<br />
그렇다면 공업제품 물가 상승률은 공업 분야에 투입되는 비중이 높은 유가의 상승과 관련이 있을까?

```python
# Crude Oil Prices

Data_CrudeOil = pd.read_csv('C:\\Users\\LG\\Desktop\\Python Term_Project\\Crude Oil Prices.csv', encoding="UTF-8")
Data_CrudeOil = Data_CrudeOil.fillna('')
Data_CrudeOil = Data_CrudeOil.drop(range(0, 10), axis = 0)
Data_CrudeOil["FRED Graph Observations"] = pd.to_datetime(Data_CrudeOil["FRED Graph Observations"])
Data_CrudeOil['Price'] = Data_CrudeOil['Price'].astype('float')
Data_CrudeOil
```

위 차트는 세계 3대 원유 중 하나인 Crude Oil(미국 서부 텍사스산 중일유)이다.<br />
Crude Oil의 경우 매주 수요일마다 지표발표가 있다는 점과 변동성이 상당히 높은 종목이기에 일반적으로 국제유가를 언급할 때에는 Crude Oil을 의미한다.<br />
위와 같은 표로 나타내어도 보는 데에는 지장이 없지만, 그래프 등으로 데이터를 표현할 때 더욱 용이하게 시각화할 수 있도록 다음 분기별로 묶인 데이터 프레임을 활용하려 한다.

```python
# Crude Oil Prices

Avg_CrudeOil = Data_CrudeOil
Avg_CrudeOil["Quarter"], Avg_CrudeOil['Year'] = (Avg_CrudeOil["FRED Graph Observations"].dt.quarter, 
                                                 Avg_CrudeOil["FRED Graph Observations"].dt.year)
pd.DataFrame(Avg_CrudeOil.groupby(['Year', 'Quarter'])['Price'].mean())
```

데이터를 그래프로 시각화할 때 X축을 설정해주어야 하기 때문에 '연도 > 분기'로 묶인 데이터 프레임을 개별의 '연도-분기' 열로 분리해야 했다.

```python
Crude = pd.DataFrame(Avg_CrudeOil.groupby(['Year', 'Quarter'])['Price'].mean())
L = []
for Year, Quarter in Crude.index:
    L.append(str(Year) + '-' + str(Quarter))
Crude["year-quarter"] = L
Crude = Crude.set_index("year-quarter")
display(Crude)
```

3개월 단위인 분기별로 묶어 보았을 때 등락 추이가 더욱 한 눈에 들어오게 된다.<br />

그 중에서도 소비자물가와 공업제품은 서로 가장 비슷한 모양을 띈다. 그렇다면 데이터 프레임의 시각화를 통해 우리의 일상 생활에 직접적인 영향을 주는 소비자물가와 앞서 살펴본 바와 같이 국제유가와 가장 관련 있다고 예상되는 공업제품을, 국제유가와 비교해보려 한다.

두 그래프를 함께 봄으로써 서로의 관계를 더욱 용이하게 살펴볼 수 있다. 공업제품의 물가는 국제유가/LNG와 더불어 2022년도 증가폭이 가장 높다.<br />
그리고 2020년도부터 꾸준한 상승세로 그래프가 이어지는 동향 또한 비슷하다. <br />

변동 추이를 함께 나타내면 다음과 같다.

```python
Data_RawMaterials['2017']=Data_RawMaterials['2017'].str.replace(pat=r',', repl=r'', regex=True)
Data_RawMaterials['2018']=Data_RawMaterials['2018'].str.replace(pat=r',', repl=r'', regex=True)
Data_RawMaterials['2019']=Data_RawMaterials['2019'].str.replace(pat=r',', repl=r'', regex=True)
Data_RawMaterials['2021']=Data_RawMaterials['2021'].str.replace(pat=r',', repl=r'', regex=True)

Data_RawMaterials = Data_RawMaterials.T

Data_RawMaterials["원유(Dubai기준)"] = Data_RawMaterials["원유(Dubai기준)"].astype('float')
Data_RawMaterials["나프타"] = Data_RawMaterials["나프타"].astype('float')
Data_RawMaterials["에틸렌"] = Data_RawMaterials["에틸렌"].astype('float')
Data_RawMaterials["프로필렌"] = Data_RawMaterials["프로필렌"].astype('float')
Data_RawMaterials["HDPE"] = Data_RawMaterials["HDPE"].astype('float')
Data_RawMaterials["PP"] = Data_RawMaterials["PP"].astype('float')

plt.figure(figsize=(10,5))
plt.grid()
plt.xlabel("연도")
plt.ylabel("원자재 가격(US/톤)")
plt.title('연도별 주요 원자재 가격')

plt.plot(Data_RawMaterials.index, Data_RawMaterials["원유(Dubai기준)"], linestyle='dashdot', marker='o')
plt.plot(Data_RawMaterials.index, Data_RawMaterials["나프타"], linestyle='dashdot', marker='o')
plt.plot(Data_RawMaterials.index, Data_RawMaterials["에틸렌"], linestyle='dashdot', marker='o')
plt.plot(Data_RawMaterials.index, Data_RawMaterials["프로필렌"], linestyle='dashdot', marker='o')
plt.plot(Data_RawMaterials.index, Data_RawMaterials["HDPE"], linestyle='dashdot', marker='o')
plt.plot(Data_RawMaterials.index, Data_RawMaterials["PP"], linestyle='dashdot', marker='o')
    
plt.legend(['원유(Dubai기준)', '나프타', '에틸렌', '프로필렌', 'HDPE', 'PP'])

plt.show()

Crude["Price"] = Crude["Price"].astype('float')
Brent["Price"] = Brent["Price"].astype('float')
LNG["Price"] = LNG["Price"].astype('float')

plt.figure(figsize=(10, 5))
plt.grid()
plt.xlabel("연도(분기)")
plt.ylabel("가격(단위: 배럴당 달러)")
plt.title('국제유가/LNG 증가율')

plt.plot(Crude.index, Crude["Price"], linestyle='dashdot', marker='.')
plt.plot(Brent.index, Brent["Price"], linestyle='dashdot', marker='.')
plt.plot(LNG.index, LNG["Price"], linestyle='dashdot', marker='.')

plt.legend(['Crude Oil', 'Brent Crude', 'LNG'])

plt.show()
```

국제유가/LNG의 상승이 석유화학분야 원자재 가격 변동에도 큰 영향 미치는 것을 한 눈에 알 수 있다. 석유화학분야 원자재의 지표는 2021년도까지 표시되어 있기 때문에 최근 2022년 그래프를 분석하는 데에는 약간의 한계가 있었지만, 국제유가가 본격적인 상승세를 보였던 2020년도부터 큰 폭으로 원자재 가격도 함께 상승했다.<br />
<br />
이제 마지막으로는 환율을 살펴보려 한다. 데이터 프레임을 살펴봤을 때 환율은 2022년이 되며 큰 폭의 변동을 나타냈다. 본 프로젝트가 중점을 두고 있는 원/달러 환율은 크게 상승하였고, 이에 따라 엔/달러 환율도 마찬가지로 상승했다. 반면 원/엔 환율은 2022년에 비교적 많이 낮은 수치를 보였다.<br />
앞서서는 국제유가와 공업제품, 석유화학분야 원자재 가격 변동의 관계성을 살펴보았고, 유가의 상승과 더불어 공업제품과 원자재 가격에서 높은 상승이 동일하게 나타났다. 이번 환율에 관련해서는 원/달러 환율의 상승이 우리 생활과 직접적인 관계가 있는 소비자물가에 미친 영향을 그래프로 살펴보려 한다.

```python
Data_KRWUSD = Data_ExchangeRate.loc[['원/달러'], :]

for col in Data_KRWUSD.columns.to_list():
    Data_KRWUSD[col] = Data_KRWUSD[col].str.replace(pat=r',', repl=r'', regex=True)

for col in Data_KRWUSD.columns.to_list():
    Data_KRWUSD[col] = Data_KRWUSD[col].astype('float')

Data_KRWUSD = Data_KRWUSD.T

plt.figure(figsize=(10,5))
plt.grid()
plt.xlabel("연도")
plt.ylabel("단위: 원(달러당)")
plt.title('원/달러 환율 추이')

plt.plot(Data_KRWUSD.index, Data_KRWUSD["원/달러"], linestyle='dashdot', marker='o')
    
plt.legend(['원/달러'])

plt.show()

Data_Price["소비자물가"] = Data_Price["소비자물가"].astype('float')

plt.figure(figsize=(10,5))
plt.grid()
plt.xlabel("연도")
plt.ylabel("증가율(%)")
plt.title('연도별 소비자물가 증가율')

plt.plot(Data_Price.index, Data_Price["소비자물가"], marker='o', color = '#1f77b4')
    
plt.legend(['소비자물가'])

plt.show()
```

원/달러 환율은 2020년부터 꾸준한 상승세를 보였다. 2020-2021 증가폭은 2019-2020 하락 폭보다 크며 2021년에 전년대비 급격한 상승을 한 뒤로 2022년에 와서는 5월에 잠깐 하락한 경우를 제외하면 지속적으로 상승했다. 원/달러 환율은 2022년 8월과 9월 사이에만 대략 100원의 상승을 하면서 달러당 1,400원을 넘어섰다.<br />
원/달러 환율과 비교하여 소비자물가를 살펴보자. 소비자물가는 2019년 전까지 하락세를 보였다. 2019-2020 기간 동안 소폭의 상승이 있었지만, 이 또한 매우 적은 수준이었다. 하지만 이후 2020년과 2021년에 접어들어 2.5%의 증가를 보여주었고, 이후 2022년에는 5.6%까지 상승하였다.<br />
이러한 동향은 원/달러 환율과 매우 흡사하다. 가장 최근 지표인 2022년을 월 단위로 분리하여 나타낸 위의 원/달러 환율 그래프는 2022년 자료를 물론 더 상세히 볼 수 있었지만, 소비자물가 증가율 또한 하락세를 보이다가 2020-2021-2022년에 대폭 상승했다는 점에서 두 지표는 영향을 주고 받으며 관계 맺고 있는 것을 알 수 있다.

이처럼 처음 시작하게 된 데이터 분석에서 라이브러리를 적절하게 사용하고 프로젝트 시작 당시 중점을 두었던 바를 분석을 통해 밝혀냈다. 그러나 데이터를 시각화함에 있어, 라이브러리와 Method 사용이 미숙했기 때문에 그래프를 단순히 그리고 눈으로만 분석하는 것에 국한되어 더욱 다양한 모습으로 시각화하지 못해 아쉬운 점도 있었다.<br />
그렇지만, 본인의 관심 주제임과 동시에 일상생활과 밀접하게 관련 있는 주제를 다루면서, 데이터 간의 관계성을 이해하며 프로젝트의 목적을 충분히 달성했다고 생각하고 세상을 보는 시야와 데이터 분석에 관련한 안목을 넓힐 수 있던 기회였다.
